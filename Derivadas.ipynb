```python
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
```

**Opci√≥n 1:** Usando np.diff

```python
x = np.linspace(-np.pi,np.pi,1000)
y = np.cos(x)

for n in range(5):
    derivada = np.diff(y) / np.diff(x)
    y = [y_i for y_i in derivada]
    x = np.linspace(np.min(x),np.max(x), len(derivada))
    plt.plot(x,y)    
```

**Opci√≥n 2: from scipy.misc import derivative**

```python
def f(x):
    return np.cos(x)

deltas = []
for i in range(20): 
    plt.plot(x, derivative(f, x0 = x, dx= 0.3, n = i , order = 6*i + 1), label = f'f{i}')
    
plt.legend(loc = 'upper right', ncol=2)
```
