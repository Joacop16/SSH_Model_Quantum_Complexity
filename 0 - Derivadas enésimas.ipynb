```python
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
```

```python
delta = 0.1

def f0(x):
    return np.cos(x)

def f1(x):
    return (f0(x + delta) - f0(x - delta))/(2*delta)

def f2(x):
    return (f1(x + delta) - f1(x - delta))/(2*delta)

def f3(x):
    return (f2(x + delta) - f2(x - delta))/(2*delta)

def f4(x):
    return (f3(x + delta) - f3(x - delta))/(2*delta)

def f5(x):
    return (f4(x + delta) - f4(x - delta))/(2*delta)

def f6(x):
    return (f5(x + delta) - f5(x - delta))/(2*delta)

def f7(x):
    return (f6(x + delta) - f6(x - delta))/(2*delta)

def f8(x):
    return (f7(x + delta) - f7(x - delta))/(2*delta)

def f9(x):
    return (f8(x + delta) - f8(x - delta))/(2*delta)

def f10(x):
    return (f9(x + delta) - f9(x - delta))/(2*delta)

def f11(x):
    return (f10(x + delta) - f10(x - delta))/(2*delta)

def f12(x):
    return (f11(x + delta) - f11(x - delta))/(2*delta)

def f13(x):
    return (f12(x + delta) - f12(x - delta))/(2*delta)

def f14(x):
    return (f13(x + delta) - f13(x - delta))/(2*delta)

def f15(x):
    return (f14(x + delta) - f14(x - delta))/(2*delta)

def f16(x):
    return (f15(x + delta) - f15(x - delta))/(2*delta)

def f17(x):
    return (f16(x + delta) - f16(x - delta))/(2*delta)

def f18(x):
    return (f17(x + delta) - f17(x - delta))/(2*delta)

def f19(x):
    return (f18(x + delta) - f18(x - delta))/(2*delta)

def f20(x):
    return (f19(x + delta) - f19(x - delta))/(2*delta)
```

```python
x = np.linspace(-np.pi,np.pi, 100)
```

```python
funciones = [f0, f1, f2, f3, f4, f5, f6, f7, f8, f9, f10, f11, f12, f13, f14]

for f in funciones:
    plt.plot(x, f(x))
```

```python
def f(x):
    return np.cos(x)


for i in range(17): 
    plt.plot(x, derivative(f, x0 = x, dx= 0.2, n = i , order = 6*i + 1), label = f'f{i}')
    
plt.legend(loc = 'upper right')
```
